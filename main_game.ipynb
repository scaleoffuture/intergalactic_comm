Добро пожаловать в Межгалактическую коммуникацию!

```python
def results(scores):
    total = 0
    for score in scores.values():
        total += score
    return total

story_log = []
violations = {}
alien_races = ["Перлы", "Ходар'Хан", "Булан-батор"]
reputation = {"Перлы": 0, "Ходар'Хан": 0, "Булан-батор": 0}

for i in range(3):
    race = alien_races[i]
    print(f"\nПереговоры с расой: {race}")
    story_log.append(f"Начало контакта с {race}.")
    
    print("Доступные действия: 1 - Мир, 2 - Ультиматум, 3 - Технологический обмен")
    choice = input("Ваш выбор: ")
    
    if choice == "1":
        reputation[race] += 20
        res = f"Вы выбрали мирный путь. {race} выражают почтение."
        print(res)
        story_log.append(res)
    elif choice == "2":
        reputation[race] -= 25
        res = f"Ультиматум привел к напряженности. {race} готовятся к атаке."
        print(res)
        story_log.append(res)
        violations[race] = "Агрессивный ультиматум" 
    elif choice == "3":
            print(f"Для обмена нужно ввести код доступа (подсказка: цифра от 1 до 3)")
            attempts = 2
            while attempts > 0:
                code = input("Введите код: ")
                if code == "3":
                    reputation[race] += 40
                    res = f"Успешный обмен с {race}. Получены новые чертежи."
                    print(res)
                    story_log.append(res)
                    break
                else:
                    attempts -= 1
                    print(f"Неверно. Осталось попыток: {attempts}")  
            if attempts == 0:
                reputation[race] -= 10
                res = f"Сбой связи с {race}. Доверие подорвано."
                print(res)
                story_log.append(res)
                violations[race] = "Сбой шифрования"
        
    else:
        reputation[race] -= 5
        res = "Вы промолчали. Дипломаты сочли это за высокомерие."
        print(res)
        story_log.append(res)
        violations[race] = "Молчание" 

final_score = results(reputation)
if final_score >= 60:
    ending = "КОНЦОВКА 1. Золотой век Галактики. Все расы объединились в союз."
elif final_score >= 30:
    ending = "КОНЦОВКА 2. Хрупкий мир. Конфликты прекращены, но доверие неполное."
elif final_score >= 0 and final_score < 30:
    ending = "КОНЦОВКА 3. Нейтралитет. Федерация сохранила границы, но не обрела союзников."
elif final_score < 0 and final_score > -30:
    ending = "КОНЦОВКА 4. Холодная война. Расы наращивают вооружение."
else:
    ending = "КОНЦОВКА 5. Галактический коллапс. Началась разрушительная война."


story_log.append(f"\nИтоговая концовка: {ending}")
story_log.append(f"Итоговые баллы: {str(reputation)}")

if violations:
    story_log.append("\nЖУРНАЛ НАРУШЕНИЙ:")
    for race, reason in violations.items():
        story_log.append(f"Раса: {race}, Причина: {reason}")

with open('intergalactic_comm.txt', 'w', encoding='utf-8') as txt_file:
    story_log.insert(0, "ИТОГОВЫЕ РЕЗУЛЬТАТЫ ИГРЫ:")
    story_log.append("\nПереговоры завершены.")

    for line in story_log:
        print(line, file=txt_file) 


print("\nРезультаты Вашей игры успешно записаны в файл .txt")
```
